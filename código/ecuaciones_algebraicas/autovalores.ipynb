```python
import sympy as sy
sy.init_printing()
a, b = sy.symbols("alpha, beta")
A = sy.Matrix([[a, b],[b, -a]])
A
```

```python
A.eigenvals()
```

```python
A.eigenvects()
```

```python
(val1, _, vec1), (val2, _, vec2) = A.eigenvects()
sy.simplify(vec1[0].T * vec2[0])
```

```python
H_s = sy.Matrix([[1, sy.Rational(1,2)], [sy.Rational(1,2), sy.Rational(1,3)]])
H_s.eigenvals()
```

```python
H_s.eigenvects()
```

```python
import scipy.linalg as la
H = la.hilbert(4)
H
```

```python
vals, vects = la.eig(H)
vals
```

```python
vects
```

```python
for i in range(len(vects)):
    for j in range(i, len(vects)):
        print(f'v[{i}] . v[{j}] = {vects[i] @ vects[j]:+.15f}')
```

```python
vals_real, vects_real = la.eigh(H)
vals_real
```

```python
import numpy as np
vals_real.sum()- np.trace(H)
```

```python
np.prod(vals_real) - la.det(H)
```
